```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def H(f, R=5e8, L=0.2, f0 = 1e9):
    return R/np.sqrt((R/L)**2+(2*np.pi*f-2*np.pi*f0**2/(f))**2)
def phase(f, R=5e8, L=1, f0=1e9):
    return np.arctan((L/R)*(2*np.pi*f0**2/f-2*np.pi*f))
```

```python
fs = np.linspace(1e6,2e9,int(1e4))
ys = H(fs)
zs = phase(fs)
plt.semilogy(fs/1e9,1-ys)
plt.axvline(1)
#plt.plot(fs/1e9,zs)
```
